# 复函数转换

```python
'''
~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~`
This draws the axis for argand diagram
~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~`
'''
plt.ylabel('Imaginary')
plt.xlabel('Real')

plt.axhline(y=0, color='black')
plt.axvline(x=0, color='black')


def argand(complex_number):
    '''
    This function takes a complex number.
    '''
    y = complex_number
    x1, y1 = [0, real(y)], [0, imag(y)]
    x2, y2 = [real(y), real(y)], [0, imag(y)]


    plt.plot(x1, y1, 'r') # Draw the hypotenuse
    #plt.plot(x2, y2, 'r') # Draw the projection on real-axis

    plt.plot(real(y), imag(y), 'bo')

r = 1
Y = [r*exp(1j*theta) for theta in linspace(0, 2*pi, 50)]
Y = array(Y)

# 画向量
#[argand(y) for y in Y]
# 画圈
plt.plot(real(Y), imag(Y), 'r')

plt.show()
```

```python
X = array([theta for theta in linspace(0, 2*pi, 50)])
Y = array([r*exp(1j*x) for x in X])

plt.plot(X, imag(Y), 'r')
plt.show()
```

# 希尔伯特变换

```python
import numpy as np
import pylab as pl
import scipy.signal as signal
from scipy import fftpack
```

```python
t = np.arange(0, 2*pi, 1 / 10000.0)
x = np.sin(t)

hx = fftpack.hilbert(x)

pl.subplot(221)
pl.plot(x, label=u"sin(t)")
pl.plot(np.real(hx), label="hilbert(sin(t))")

pl.title(u"Hilbert Transform")
pl.legend()
```
