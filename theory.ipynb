# 交易和市场
交易：
- 左侧交易：价值投资，依赖对数千种金融衍生品的动态的基本面分析（市场beta，量化难度更高），难度较高，回撤较大，适合大资金，专业的机构投资者（代表人物巴菲特）
- 右侧交易：趋势投机，依赖于对趋势的判断（技术分析），难度较低（只需要回测，挖掘因子alpha），回撤较小，适合小资金，有专业知识的散户投资者（代表人物利弗莫尔）
- 随着资金规模的增大，交易点（和策略）将不得不向左移，来保证回报率
- ”斩断亏损，让利润奔跑“ 的策略，天然更容易被右侧交易实现

量化：
- p quant:(physical prob measure) 预期未来价格，alpha mining，常见于买/卖方(机构)自营交易部和散户交易者，可以左/右侧交易
- q quant:(q measure)发布复杂对冲衍生品，赚取手续费，为市场融资提供流动性，提升市场效率，常见于卖方机构

### Utility functions and portfolio choice

- VNM: [Von Neumann-Morgenstern Utility Theorem](https://en.wikipedia.org/wiki/Von_Neumann%E2%80%93Morgenstern_utility_theorem) 
- [Expected Utility Hypothesis](https://en.wikipedia.org/wiki/Expected_utility_hypothesis)

$$
\color{Green}
\begin{align*}
&\forall \text{ random } \widetilde{w} \text{ with mean } \overline{w}\\
&\text{utility function } u \text{ :}\\
    &\begin{cases}
    &\widetilde{w_{1}} \text{ is as prefered: }  E[u(\widetilde{w_{1}})] \ge E[u(\widetilde{w_{2}})]\\
    &u \text{ is (weakly) risk averse: } u(\overline{w}) \ge E[u(\widetilde{w})] \text{ (concave)}\\
    &\text{ for } w, u \text{ is unique for all monotone } u\\
    &\text{ for } \widetilde{w}, u \text{ is unique for all affine monotone } u\\
    &(\exists \ a, b(> 0), u_{1}, u_{2} \text{ are affine for }  u_{1}(w) = a + bu_{2}(w))
    &\end{cases}\\

&\end{align*}\\
$$

- high curvature of utility function at a wealth level, measured by coefficient of absolute risk $\alpha(w) = -\frac{u''(w)}{u'(w)}$, represents high aversion to risk
- for H investors, the aggregated absolute risk aversion is $\frac{1}{\sum^{H}_{h=1}{1/\alpha_h}}$
- the amount of money investor willing to pay to avoid taking a gamble at current wealth level (rational investor behavior) can be calculated. different for large and small gambles
- multiple utility functions can be found, for example, by assuming a constant risk aversion coefficient at all wealth level
- for a investor with concave $u$, $ E[u(\widetilde{w}+\widetilde{\epsilon})] \le E[u(\widetilde{w})]$ for $ E[\widetilde{\epsilon}|\widetilde{w}] = 0 \ (E[\widetilde{\epsilon}] = 0, cov(\widetilde{\epsilon},\widetilde{w}) = 0)$, meaning that investor of this type would ALWAYS dislike mean independent noise(gamble) epsilon at current wealth level. 

$$
\color{Green}
\begin{cases}
    &\max_{\theta_{i}}{E[u(\widetilde{w})]}\\
    &\widetilde{w} = \widetilde{y} + \sum^{n}_{i=1}{\theta_{i}\widetilde{x_{i}}}\\
    &\sum^{n}_{i=1}{\theta_{i}p_{i}} = w_{0}\\
\end{cases}\\
\Rightarrow
\begin{cases}
  &\forall{i} \ E[\frac{u'(\widetilde{w})}{\gamma} \frac{\widetilde{x_{i}}}{p_{i}}] = 1\\
  &E[u'(\widetilde{w})(\widetilde{R_{i}}-\widetilde{R_{j}})] = 0

\end{cases}
$$

- a rational investor should aim for this optimization problem:
  - $\theta_{i}$ is the volume of share units held for asset i
  - $p_{i}$ is the initial unit share price and $\widetilde{x_{i}}$ is the random unit share price at the end of the period
  - $\widetilde{y}$ is for added generality for possible personal endowment at the end of the period (e.g. income)
- by [Lagrange multiplier theorem](https://en.wikipedia.org/wiki/Lagrange_multiplier), the lagrangian (for unique multiplier $\gamma$) of the problem is 
$
E[u(\widetilde{y} + \sum^{n}_{i=1}{\theta_{i}\widetilde{x_{i}}})] - \gamma (\sum^{n}_{i=1}{\theta_{i}p_{i}} - w_{0}) = 0
$
  - the first order condition tells us that two random variables $u'(\widetilde{w})$ (marginal utility evaluated at optimal wealth level (1 for risk-neutral investors)) and $(\widetilde{R_{i}}-\widetilde{R_{j}})$ (each excess return for a zero-cost portfolio of holding asset i by shorting asset j) are orthogonal
  - if by shorting asset j and adding asset i (zero-cost) I can increase or decrease utility return, then current portfolio is definitely not optimal
  - if a market has a risk free asset with free-risk return $R_{f}$ this condition also holds for $R_{j} = R_{f}$

### SDF: Stochastic Discount Factor and [Pricing Kernel](https://en.wikipedia.org/wiki/Stochastic_discount_factor)
- in previous optimal portfolio condition, we have: $\forall{i} \ E[\frac{u'(\widetilde{w})}{\gamma} \widetilde{x_{i}}] = p_{i}$
- this means something like $\forall{i} \ E[\widetilde{m} \widetilde{x_{i}}] = p_{i}$ could exist for all portfolios, where $\widetilde{m}$ is the "pricing kernel" (SDF)
- $\exist \text{ no arbitrage } \Rightarrow \text{law of one price}$
  - law of one price is a weaker condition than absence of arbitrage opportunities
  - it simply states that any two portfolio with the same payoff must have the same cost. If it fails, investor can arbitrage on portfolio with less cost

$$
\color{Green}
\text{Law of One Price: for any two portfolio } \theta, \psi: \\
\sum^{n}_{i=1}{\theta_{i}\widetilde{x_{i}}} = \sum^{n}_{i=1}{\psi_{i}\widetilde{x_{i}}} 
\Rightarrow
\sum^{n}_{i=1}{\theta_{i}\widetilde{p_{i}}} = \sum^{n}_{i=1}{\psi_{i}\widetilde{p_{i}}} 
$$

- here we state:
  - LOP implies:
    - current price p is a function of future price $\widetilde{x}$
    - $p(\widetilde{x})$ is a linear functional: $ p(\alpha_{1}\widetilde{x_{1}}+\alpha_{2}\widetilde{x_{2}}) = \alpha_{1}p(\widetilde{x_{1}})+\alpha_{2}p(\widetilde{x_{2}}) $ (otherwise arbitrage exists)
  - followed by [Riesz_Representation_Theorem](https://en.wikipedia.org/wiki/Riesz_representation_theorem): bounded linear functionals (current price p) in a Hilbert space have an inner product representation
  - (recall that if there is no arbitrage, x(initial price)=E[mX] is in the cone of $\widehat{X}$, m is strictly positive)
$$
\color{Green}
\text{Law of One Price } \Leftrightarrow \text{current price p is a linear functional on asset plane spanned by future price x}\\
\text{Law of One Price } \Leftrightarrow \exist \text{ SDF } \widetilde{m} : E[\widetilde{m} \widetilde{x_{i}}] = p_{i}\\
\exist \text{ no arbitrage } \Leftrightarrow \exist \text{ strictly positive SDF } \widetilde{m} : E[\widetilde{m} \widetilde{x_{i}}] = p_{i}\\
\text{ complete markets } \Leftrightarrow \text{ SDF is unique }\\
$$
- consider a risk free asset, SDF must also hold for this asset, which implies that $E[\widetilde{m}] = \frac{1}{R_f}$ , risk premium of a portfolio is:
$$
\color{Green}
E[\widetilde{R}] - R_{f} = R_f(E[\widetilde{m} \widetilde{R}]-cov(\widetilde{m}, \widetilde{R})) - R_f = -R_f cov(\widetilde{m}, \widetilde{R})
$$

- Geometric interpretation:
  - [Linear Regression](https://en.wikipedia.org/wiki/Linear_regression): consider n samples scattered in a p-dimensional scalar field along a line: $\{y_i, x_{i1}, ..., x_{ip}\}_{i=1}^{n}$
  - model it by 
  $$
  \begin{cases}
  &y_i = \beta_0 + \beta_1 x_{i1} + ... + \beta_p x_{ip} + \widetilde{\epsilon_i}\\
  &(\mathbf{y}_{n \times 1} = \mathbf{X}_{n \times (1+p)} \mathbf{\beta}_{(1+p) \times 1} + \mathbf{\epsilon}_{n \times 1})\\
  \end{cases}
  $$
  - by minimizing residue square, we can get [Ordinary Least Square](https://en.wikipedia.org/wiki/Ordinary_least_squares) (or [Generalized least squares](https://en.wikipedia.org/wiki/Generalized_least_squares)): 
  $$
  \begin{align*}
  &\widehat{\beta}= \text{argmin}_{b\in \mathbb{R}^{p}}{\sum^{n}_{i=1}{(y_i-x_i^{T} b)^2}} = \text{argmin}_{b\in \mathbb{R}^{p}}{(y-\mathbf{X}b)^{T} (y-\mathbf{X}b)}\\
  &=(X^{T}X)^{-1}X^{T}y\\
  &y_{proj}=X(X^{T}X)^{-1}X^{T}y\\
  \end{align*}\\
  $$
  ![Linear Regression](figure/LR.png)

- we have established that LOP implies current price is in asset plane spanned by future price, it remains to show that SDF ($E[\widetilde{m} \widetilde{x_{i}}] = p_{i}$) $\widetilde{m}$ equals to the unique stochastic discount factor in the asset plane spanned by future price ($\widetilde{m_{proj}}$) + the component orthogonal to asset plane spanned by future price ($\widetilde{\epsilon}$):
  - if an element is in the asset plane spanned by future price, it is future payoff of a portfolio
  - sketch of proof: 
    - if we suppose a component $\epsilon$ of SDF is orthogonal to the future asset plane $E[\widetilde{X}(\widetilde{m}-\widetilde{X}'\theta-a)] = 0$, where $\widetilde{X}'\theta+a$ is on asset plane（risky and risk-free asset).
    - This means that SDF can be modeled by Linear regression models like OLS/GLS.
$$

\color{Green}
  \begin{align*}
    \widetilde{m}
    & = \theta'\widetilde{X} + \widetilde{\phi} (\text{assume } \phi \text{ is orthogonal to X and show }\theta\text{ is unique})\\
    & = \underbrace{E[\widetilde{X}\widetilde{X}']^{-1}E[\widetilde{X}\widetilde{m}]'\widetilde{X}}_{\text{always unique but theta only unique if Cov(X) is invertible(l.i.)}} +  \widetilde{\phi}\\
    & = p'E[\widetilde{X}\widetilde{X}']^{-1}\widetilde{X} +  \widetilde{\phi}\\
    & = \widetilde{m_{proj}} + \widetilde{\phi}\\
    & = A + B'\widetilde{X} + \widetilde{\epsilon}
      \begin{cases}
        &A = \bar{m} - B\bar{X}\text{ (may not unique) }\\
        &B = Cov(\widetilde{m}, \widetilde{X})Cov(\widetilde{X})^{-1} \text{ (may not unique) }\\
        &\widetilde{\epsilon} = (\widetilde{m} - \bar{m}) - B(\widetilde{X} - \bar{X})\\
        &E[\widetilde{\epsilon}] = 0\\
        &E[\widetilde{\epsilon}\widetilde{X}'] = 0\\
      \end{cases}\\
    & = \widetilde{m_{proj}} + \widetilde{\epsilon}\\
  \end{align*}\\
$$

- Hansen-Jagannathan bound:
- 
  - no risk-free asset: 
$$
\text{stdev}(\widetilde{m}) \ge \frac{|E[\widetilde{m}]E[\widetilde{R}]-1|}{\text{stdev}(\widetilde{R})}
$$
- 
  - with risk-free asset: 
$$
\text{stdev}(\widetilde{m}) \ge \frac{E[\widetilde{R}]-R_f}{R_f \text{stdev}(\widetilde{R})}
$$

### [General Equilibrium Theorem](https://web.stanford.edu/~jdlevin/Econ%20202/General%20Equilibrium.pdf)

- Pareto Optima: consider H investor with utility functions $u_h$ and their initial allocations, the reallocation of the random aggregate market wealth is at Pareto Optima if there exist no other allocation than the current one such that each investor can be at least as better off. (this means we cannot improve a investors' interests without harming another one)
- Competitive Equilibria: is a set of prices $(p_1, ..., p_n)$ and portfolios $(\theta_1, ..., \theta_n)$ such that:
  - market is clear: for each commodity, the sum of this commodity type from each investor under current allocation is the same as initial total volume of that commodity
  - each investor(agent) try to optimize portfolio, subject to initial new wealth and possible endowments at the end
$$
\begin{cases}
&\text{market is clear: } (\forall i) \sum^{H}_{h=1}{\theta_{hi}} = \bar{\theta_i}\\
&(\forall h) \text{ max } E[u_h(\widetilde{w_h})] 
  \begin{cases}
  &\text{fixed initial wealth: } \sum^{n}_{i=1}{\theta_{hi}p_i} = \sum^{n}_{i=1}{\bar{\theta_{hi}}p_i}\\
  &(\forall \omega) \widetilde{w_h}(\omega) = \widetilde{y_h}(\omega) + \sum^{n}_{i=1}{\theta_{hi}\widetilde{x_i}(\omega)}\\
  \end{cases}
\\
\end{cases}\\
$$
- complete markets: a security market is complete if for any random wealth with state $\omega$ there exists a portfolio $\widehat{\theta}$ to span it (addition of other financial derivatives also helps market completion)
$$
(\forall \omega) \sum^{n}_{i=1}{\theta_i \widetilde{x_i}(\omega)} = \widetilde{w}(\omega)
$$

- Here we state: (without proof)
  - if an allocation is at both competitive equilibrium and Pareto optimal, there is a representative investor who would optimize and have control over all assets and endowments of the original economy
  - for a competitive equilibrium to be Pareto optimal, it is sufficient that market is complete and investors all have linear risk tolerance $\tau(w) = A + Bw$ with B as constant among all investors 


### Mean-Variance Frontier (minimizing portfolio variance)

- consider only risky assets:
$$
\color{Green}
\begin{cases}
    &\pi \text{ is the portfolio vector consists of n assets}\\
    &{\textstyle \sum_{\pi}} \text{ is the } n\times n \text{ covariance matrix consists of } cov(\widetilde{R_{i}, R_{j}}) \text{ as its (i, j)-th element} \\
    &R_{i} \text{ is the return of risky asset i}\\
    &\mu \text{ is the vector of expected returns of assets}\\
    &\mu_{p} \text{ is the target return we aim at}\\
\end{cases}\\
\min{\pi'{\textstyle \sum_{\pi}}} \text{ subject to } \mu'\pi = \mu_{p} \text{ and } \textbf{1}'\pi = 1\\
\begin{align*}
    \text{Lagrange }\Rightarrow \pi
    & = \lambda\pi_{\mu}+(1-\lambda)\pi_{1}
    \begin{cases}
        &\lambda = \frac{\mu_{p}-\mu'\pi_{1}}{\mu'(\pi_{\mu}-\pi_{1})}\\
        &\pi_{\mu} = \frac{1}{\textbf{1}'{\textstyle \sum^{-1}}\mu}{\textstyle \sum^{-1}}\mu\\
        &\pi_{1} = \frac{1}{\textbf{1}'{\textstyle \sum^{-1}}\textbf{1}}{\textstyle \sum^{-1}}\textbf{1}\\
    \end{cases}\\
    & = \phi_{a}\pi_{a}+(1-\phi_{a})\pi_{b}
    \begin{cases}
        &\pi_{a} = \lambda_{a}\pi_{\mu}+(1-\lambda_{a})\pi_{1}\\
        &\pi_{b} = \lambda_{b}\pi_{\mu}+(1-\lambda_{b})\pi_{1}\\
        &\phi_{a} = \frac{\lambda-\lambda_{b}}{\lambda_{a}-\lambda_{b}}\\
    \end{cases}
\end{align*}\\
\begin{align*}
    \text{Var[any Frontier portfolio}]
    & = [\lambda\pi_{\mu}+(1-\lambda)\pi_{1}]'{\textstyle \sum}[\lambda\pi_{\mu}+(1-\lambda)\pi_{1}]\\
    & = \frac{A-2B\mu_{p}+C\mu_{p}^2}{AC-B^{2}}
    \begin{cases}
        & \text{ const } A = \mu'{\textstyle \sum^{-1}\mu}\\
        & \text{ const } B = \mu'{\textstyle \sum^{-1}\textbf{1}}\\
        & \text{ const } C = \textbf{1}'{\textstyle \sum^{-1}\textbf{1}}\\

    \end{cases}
    \\
\end{align*}

$$
- this means that for investor who have mean-variance preference, any two portfolios on the mean-variance-frontier can span the whole fronter by choosing different weights
- (standard deviation, expected return) plot has a locus of a hyperbola

- More generally, consider risky assets with a risk free asset:
$$
\color{Green}
\min{\pi'{\textstyle \sum_{\pi}}} \text{ subject to } (\mu-R_{f}\textbf{1})'\pi = \mu_{p}-R_{f}\\
\Rightarrow \pi = \frac{\mu_{p}-R_{f}}{(\mu-R_{f}\textbf{1})'{\textstyle \sum^{-1}}(\mu-R_{f}\textbf{1})}{\textstyle \sum^{-1}}(u-R_{f}\textbf{1})\\
\text{stdev[any Frontier portfolio]} = \frac{\overbrace{|\mu_{p}-R_{f}|}^{\text{Risk Premium}}}{\underbrace{\sqrt{{(\mu-R_{f}\textbf{1})'{\textstyle \sum^{-1}}}(\mu-R_{f}\textbf{1})}}_{\text{Sharp Ratio}}}\\
$$
- unlike risky assets only, risk-free asset and any other non-purely-risk-free asset portfolio on the portfolio frontier span the whole portfolio frontier using different weights

### Returns
- assume LOP holds (thus SDF exists):
$$
\color{Green}
\begin{align*}
    p
    &= E[\widetilde{m}\widetilde{X}] = Cov(\widetilde{m}, \widetilde{X}) + E[\widetilde{m}]E[\widetilde{X}]\\
    &\Rightarrow 1 = Cov(\widetilde{m}, \widetilde{R}) + E[\widetilde{m}]E[\widetilde{R}]\\
    & 
    \begin{align*}
        \Rightarrow E[\widetilde{R}]
        &= \frac{1}{E[\widetilde{m}]}-\frac{1}{E[\widetilde{m}]}Cov(\widetilde{m}, \widetilde{R})\\
        &= R_{f}-R_{f}Cov(\widetilde{m}, \widetilde{R}) \text{ if a risk-free asset exists (m also hold) and LOP holds}\\
    \end{align*}

    \\
\end{align*}
$$
- recall that since at least one SDF exists and let $\widetilde{m_{p}}$ be one of the allowed SDF that is the unique projection of all SDFs on the asset span (it is the payoff of certain portfolio which must exist) and itself is also one of the SDF: let's consider that specific portfolio:
$$
\color{Green}
p = E[\widetilde{m}\widetilde{X}] = E[\widetilde{m_{p}}\widetilde{X}] = E[\widetilde{m_{p}}^{2}]\\
\widetilde{R} = \frac{\widetilde{X}}{E[\widetilde{m}\widetilde{X}]}\\
\Rightarrow \underbrace{\widetilde{R_{p}}}_{\text{unique}} = \frac{\widetilde{m_{p}}}{E[\widetilde{m_{p}}^{2}]}
\begin{cases}
    &\text{for the projection portfolio}\\
    &\text{if that SDF exists}\\
\end{cases}\\
$$
- to prove that any return can be decomposed and the decomposition is useful, it suffices to show that residue item has 0 mean and is orthogonal to other components, and other variables are unique:

- returns can be influenced by several actions: assets price change overtime while holding certain portfolio, and dynamically changing the portfolio without endowment, by shorting and longing different assets by equal amount:
  - this action may or may not be beneficial
  - thus the excess returns are a finite-dimensional linear subspace of asset price/return space:  $a*(\widetilde{R_{i}}-\widetilde{R_{j}})$
  - consider projection of return $\mathbf{1}$ onto this subspace as $\widetilde{e_{p}}$
$$
\color{Green}
\widetilde{x} \in \text{excess return subspace} \in \text{return space}:\\
E[\widetilde{e_{p}}\widetilde{x}] = E[(1-\underbrace{\widetilde{\phi}}_{\text{orthogonal to excess return subspace}})\widetilde{x}] = E[\widetilde{x}]
$$

- consider $\widetilde{e_{p}}$ as the Riesz representation of the bounded linear functional $\widetilde{x} -> E[\widetilde{x}]$ on the space of excess returns(payoffs of zero-cost portfolio)
$$
\color{Green}
\text{assume: } \widetilde{R} = \underbrace{a\widetilde{R_{p}}+b\widetilde{e_{p}}}_{
        \text{orthogonal projection of R on span of } \widetilde{R_{p}}, \widetilde{e_{p}}\\
    }

[sketch of proof](https://arxiv.org/pdf/1912.01091.pdf) (Theorem 3.1 Lemma 3.2):

- about probability measure: 
  - a Banach space is a vector space that is complete (e.g. no missing holes, every Cauchy Sequence of its points have a limit also in it) and with definition of a norm (thus can calculate distance). Thus $\mathbf{R}^{m}$ is a typical banach space
  - a functional space is also a vector space, of infinite dimensions
  - a banach algebra defined on a set can be a bounded functional space on this set, which is a vector space, which can have its functional dual space of all linear functionals, which can be proven to have the same dimensions as the original functional(vector) space, which is infinite
  - A linear functional f maps a linear subspace (e.g. a convex cone) of a vector space to a scalar field (real/complex), such that $f(\alpha x + \beta y) = \alpha f(x) + \beta f(y)$
  - $B(\Omega)$ can be thought as function that maps subset of $\Omega$ (compact or not) to "Area" or "Volume", a Probability Measure is a function defined in $B(\Omega)$'s dual space of linear functionals to map "Volume" to Probability, such that for Random Variables A, B of the same measure $\textit{P}$ : $\textit{P}(aX + bY) = a\textit{P}(X)+b\textit{P}(Y)$
  - a discrete space can also be compact, as long as it's finite
  - a compact market is never possible, it needs negligible transaction costs and also perfect information

- since we assume X is bounded (a reasonable assumption for financial market), $\Omega$ is a bounded set usually centered near(or not) x, there must exist a smallest closed cone $\{ <X, \Pi> : \Pi \in \textit{P} \}$ containing the range of X: $\Omega$
- why it is a cone?
  - X is a function that maps a vector x (a linear subspace of $\mathbf{R}^{m}$) to a set of vectors $\Omega$ (sample space), note that X is not a linear functional
  - $\Pi$ is a function that maps a Banach Algebra in $\Omega$ to a scalar probability, it is a linear functional
  - the inner product of these 2 functions are an integral of 2 functions (which would result in a vector * scalar = vector) over $\Omega$. Note that for inner product to be defined between functions, at least one function needs to resides in Banach Space, which in this case, is measure $\Pi$, as it is a linear functional of a Banach Space(Algebra)
  - if $\Pi$ is not constrained, $\{ <X, \Pi> : \Pi \in \textit{P} \}$ is naturally the smallest cone containing $\Omega$
  - if market is Arbitrage-Free, the price deflator is simply a measure to find x in a cone, if market is complete, the measure is unique 

- if x belong to the cone, the position set up cost cannot be negative (contrapositive)
- it is a well known result(non-trivial) that since a cone is closed and convex and x is not in the cone, there must be an arbitrage cone (proof omitted)
- let $x^{*}$ be the nearest point in the cone containing $\Omega$ to x, $\gamma = x^{*} - x$ must be in the arbitrage cone

### Multi-Period Model

$$
\color{Green}
\begin{cases}
    &(t_{j})_{0 \le j \le n}\text{: increasing time sequence}\\
    &(\mathcal{A}_{j})_{0 \le j \le n}\text{: }
        \begin{cases}
        &\text{sequence of Algebras}\\
        &\text{on set of possible outcomes of } \Omega\\
        &\text{represent information available at time}\\
        &\text{cardinality of } \mathcal{A}_{0} \text{ is finite}\\
        &\mathcal{A}_{j} \text{ is increasing (includes previous algebras)}\\
        \end{cases}\\
    &(X_{j})_{0 \le j \le n}\text{: }
        \begin{cases}
        &\text{sequence of bounded } \mathbf{R}^{m} \text{valued functions}\\
        &\mathcal{A} \text{ measurable} \\
        &\text{represent prices of m instruments}\\
        \end{cases}\\
    &(C_{j})_{1 \le j \le n}\text{: }
        \begin{cases}
        &\text{sequence of bounded } \mathbf{R}^{m} \text{valued functions}\\
        &\mathcal{A}_{j} \text{ measurable} \\
        &\text{represent cash flow holding 1 share over all previous period}\\
        \end{cases}\\
    &(\Gamma_{j})_{0 \le j \le n}\text{: }
        \begin{cases}
        &\text{sequence of bounded } \mathbf{R}^{m} \text{valued functions}\\
        &\mathcal{A}_{j} \text{ measurable} \\
        &\text{represent (trading strategy) amount of security purchase at period}\\
        \end{cases}\\

    &\begin{cases}
    &\text{position: } \Xi_{j} = \Gamma_{0} + ... + \Gamma_{j} \text{ (accumulated trade volume over time)}\\
    &\text{closed out: } \Xi_{j} = 0\\
    \end{cases}\\
        
    &(A_{j}) \text{: }
        \begin{cases}
        &0 \le j \le n\\
        &C_{0} = 0 \\
        &A_{j} = \Xi_{j-1} \cdot C_{j} - {\Gamma_{j} \cdot X_{j}} \text{ : account money made at } t_{j}\\
        &\underbrace{\text{previous net position in volume}}_{\Xi_{j-1}} * \underbrace{\text{price inflation (in dollar) over last period}}_{C_{j}}\\
        &\underbrace{\text{new purchase in volume}}_{\Gamma_{j}} \text{ at } \underbrace{\text{current price in dollar}}_{X_{j}}\\
        \end{cases}\\
\end{cases}\\

\Rightarrow
\begin{cases}
    &\exists \text{ Arbitrage }
    \Leftrightarrow
    \begin{cases}
    &\exists \ \Gamma_{j} \text{ making strictly positive on initial trade}\\
    &\text{and non-negative amounts until closing-out}\\
    \end{cases}\\

    &\exists \text{ no Arbitrage } \Leftrightarrow 
        \begin{cases}
        &\exists \ \oplus\Pi_{i} \text{ s.t. vector-valued measure：}\\
        &\begin{split}
            X_{i} \cdot \Pi_{i} 
            & = (C_{i+1}+X_{i+1}) \cdot \Pi_{i+1}|\mathcal{A}_{i} \ , \ 0 \le i <n\\
            & = \sum_{i<j<k}{C_{j+1} \cdot \Pi_{j}|\mathcal{A}_{i} + (C_{k} + X_{k}) \cdot \Pi_{k}|\mathcal{A}_{i}} \ , \ i<k\\
        \end{split}\\

        &:\underbrace{\text{expected current price}}_{X_{i} \cdot \Pi_{i}}\\
        &:\underbrace{\text{expected next liquidized cash + remaining stock price}}_{(C_{i+1}+X_{i+1}) \cdot \Pi_{i+1}}\\
        &{(X_{i} \cdot \Pi_{i})}_{i\ge0} \text{ being a martingale}
        \end{cases}\\

\end{cases}
$$

sketch of proof:
- for multi-period model, at time t, the last transaction is a purchase. rather than clearing out like the single period model
- for the end of strategy (closing-out), the last transaction is also a selling
- conditional expectation:
  - function $\Omega \rightarrow \mathbf{R}^{m} \text{: } \ Y = \mathit{E}[X|\mathcal{A}]$ for random variable X 
  $$ 
  \Leftrightarrow
  \begin{cases}
  &\text{Y is } \mathcal{A} \text{ measurable }\\
  &\int_{A}{Y dP} = \int_{A}{X dP} \  \forall A \in \mathcal{A}
  \end{cases}
  $$
  - for random variable A, B, conditional expectation of A given B is a random variable of B
- $X_{i}\Pi_{i}$ : is inner product(integral) over $\mathbf{R}^{m}$, so it's expected value of random variable
- note that for a price vector space describing the current payments, there could also be an equivalent pay off space describing the future payments
- this proves that if a market is effective (arbitrage-free in probabilistic sense of possible next prices), and no cash flow (liquidizing stocks), given observation at time t, the expected price later should be the same as now, meaning that market price should be a martingale sequence
